```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generate the 2-class moon dataset
X, y_true = make_moons(n_samples=1000, noise=0.05, random_state=42)

# Visualize the true labels
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=50, alpha=0.7)
plt.title('2-Class Moon Dataset (True Labels)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='True Class')
plt.grid(True)
plt.show()
```

```python
# Apply DBSCAN
# You might need to tune eps and min_samples for optimal results
dbscan = DBSCAN(eps=0.3, min_samples=4)
clusters = dbscan.fit_predict(X)

print(f"Number of clusters found by DBSCAN: {len(np.unique(clusters[clusters != -1]))}")
print(f"Number of noise points (labeled -1): {np.sum(clusters == -1)}")

# Visualize the DBSCAN results
plt.figure(figsize=(10, 7))
# Plot noise points in black
plt.scatter(X[clusters == -1, 0], X[clusters == -1, 1], c='black', marker='x', s=100, label='Noise (Outliers)')
# Plot clustered points
plt.scatter(X[clusters != -1, 0], X[clusters != -1, 1], c=clusters[clusters != -1],
            cmap='plasma', s=50, alpha=0.7, label='Clusters')

plt.title('DBSCAN Clustering Results on 2-Class Moon Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='DBSCAN Cluster')
plt.legend()
plt.grid(True)
plt.show()
```
